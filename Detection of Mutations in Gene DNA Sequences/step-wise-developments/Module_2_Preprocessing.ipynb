Function to get the Reference Gene seqeunces from the COSMIC API

```python
from string import Template
import requests

BASE_URL = 'https://cancer.sanger.ac.uk/cosmic/'
SEQ_URL = Template('sequence?ln=$gene&type=$_type')

def get_gene_seq(gene, _type):
  URL = BASE_URL + SEQ_URL.substitute(gene=gene,_type=_type)
  print(f'GET ---> {URL}')
  request = requests.get(url = URL)
  sequence = request.text
  sequence = sequence.split('\n')
  sequence = ''.join(sequence[1:len(sequence)-1])
  print(sequence)
  print(len(sequence))
  return sequence;

gene_seq = dict()
genes = list(selected_genes['GENE_NAME'].unique())
for gene_name in genes:
  seq = get_gene_seq(gene_name,'cdna')
  gene_seq[gene_name] = seq
```

**Function to create Mutated Genes for Substitutions**

```python
import re

# regex for substitution
regex = '^c\.\d+[A-Z]>[A-Z]$|^c\.\d+_\d+delins[A,G,C,T]+$'

# generate mutated gene DNA sequence - for substitution - with 1 mutation
def get_mutated_gene_sequence_sub(gene, substitution):
  if gene in gene_seq.keys():
    print(substitution)
    if re.search(regex,substitution):
      numbers = list(map(int,re.findall(r'\d+',substitution)))
      temp = list(gene_seq[gene])
      if len(numbers) == 1:
        substitute_index=numbers[0]-1
        temp[substitute_index] = substitution.split('>')[1]
        return ''.join(list(temp))
      if len(numbers) == 2:
        temp[numbers[0]-1]=substitution[-2]
        temp[numbers[1]-1]=substitution[-1]
        return ''.join(list(temp))
      raise Exception("Something went wrong")
    
    raise Exception("substitution is invalid")
  raise Exception(f"Unknown gene.\nGenes: %s",','.join(list(genes.keys())))


# generate mutated gene DNA sequence - for substitution - with more than 1 mutations
def get_mutated_gene_sequence_sub_multiple(substitution,sequence):
  print(substitution)
  if re.search(regex,substitution):
    numbers = list(map(int,re.findall(r'\d+',substitution)))
    temp = list(sequence)
    if len(numbers) == 1:
      substitute_index=numbers[0]-1
      temp[substitute_index] = substitution.split('>')[1]
      return ''.join(list(temp))
    if len(numbers) == 2:
      temp[numbers[0]-1]=substitution[-2]
      temp[numbers[1]-1]=substitution[-1]
      return ''.join(list(temp))
    raise Exception("Something went wrong")
  
  raise Exception("substitution is invalid")
```

**Method to Simulate Errors in the Sequence (to be used inside Generate_Mutated_Sequence function)**

```python
import random

def sim_error(sequence,max_error_percentage):
  length = len(sequence)
  seq_specific_error = random.uniform(0,max_error_percentage)
  substitute_count = round(length*seq_specific_error)

  s = list(sequence)
  while(substitute_count>0):
    n = random.randint(0,length-1)
    s[n] = random.choice(["A","C","G","T"])
    substitute_count -= 1
  
  return "".join(s)

# error = sim_error('ATGGCGGCGCTGAGCGGTGGCGGTGGTGGCGGCGCGGAGCCGGGCCAGGCTCTGTTCAACGGGGACATGGAGCCCGAGGCCGGCGCCGGCGCCGGCGCCGCGGCCTCTTCGGCTGCGGACCCTGCCATTCCGGAGGAGGTGTGGAATATCAAACAAATGATTAAGTTGACACAGGAACATATAGAGGCCCTATTGGACAAATTTGGTGGGGAGCATAATCCACCATCAATATATCTGGAGGCCTATGAAGAATACACCAGCAAGCTAGATGCACTCCAACAAAGAGAACAACAGTTATTGGAATCTCTGGGGAACGGAACTGATTTTTCTGTTTCTAGCTCTGCATCAATGGATACCGTTACATCTTCTTCCTCTTCTAGCCTTTCAGTGCTACCTTCATCTCTTTCAGTTTTTCAAAATCCCACAGATGTGGCACGGAGCAACCCCAAGTCACCACAAAAACCTATCGTTAGAGTCTTCCTGCCCAACAAACAGAGGACAGTGGTACCTGCAAGGTGTGGAGTTACAGTCCGAGACAGTCTAAAGAAAGCACTGATGATGAGAGGTCTAATCCCAGAGTGCTGTGCTGTTTACAGAATTCAGGATGGAGAGAAGAAACCAATTGGTTGGGACACTGATATTTCCTGGCTTACTGGAGAAGAATTGCATGTGGAAGTGTTGGAGAATGTTCCACTTACAACACACAACTTTGTACGAAAAACGTTTTTCACCTTAGCATTTTGTGACTTTTGTCGAAAGCTGCTTTTCCAGGGTTTCCGCTGTCAAACATGTGGTTATAAATTTCACCAGCGTTGTAGTACAGAAGTTCCACTGATGTGTGTTAATTATGACCAACTTGATTTGCTGTTTGTCTCCAAGTTCTTTGAACACCACCCAATACCACAGGAAGAGGCGTCCTTAGCAGAGACTGCCCTAACATCTGGATCATCCCCTTCCGCACCCGCCTCGGACTCTATTGGGCCCCAAATTCTCACCAGTCCGTCTCCTTCAAAATCCATTCCAATTCCACAGCCCTTCCGACCAGCAGATGAAGATCATCGAAATCAATTTGGGCAACGAGACCGATCCTCATCAGCTCCCAATGTGCATATAAACACAATAGAACCTGTCAATATTGATGACTTGATTAGAGACCAAGGATTTCGTGGTGATGGAGGATCAACCACAGGTTTGTCTGCTACCCCCCCTGCCTCATTACCTGGCTCACTAACTAACGTGAAAGCCTTACAGAAATCTCCAGGACCTCAGCGAGAAAGGAAGTCATCTTCATCCTCAGAAGACAGGAATCGAATGAAAACACTTGGTAGACGGGACTCGAGTGATGATTGGGAGATTCCTGATGGGCAGATTACAGTGGGACAAAGAATTGGATCTGGATCATTTGGAACAGTCTACAAGGGAAAGTGGCATGGTGATGTGGCAGTGAAAATGTTGAATGTGACAGCACCTACACCTCAGCAGTTACAAGCCTTCAAAAATGAAGTAGGAGTACTCAGGAAAACACGACATGTGAATATCCTACTCTTCATGGGCTATTCCACAAAGCCACAACTGGCTATTGTTACCCAGTGGTGTGAGGGCTCCAGCTTGTATCACCATCTCCATATCATTGAGACCAAATTTGAGATGATCAAACTTATAGATATTGCACGACAGACTGCACAGGGCATGGATTACTTACACGCCAAGTCAATCATCCACAGAGACCTCAAGAGTAATAATATATTTCTTCATGAAGACCTCACAGTAAAAATAGGTGATTTTGGTCTAGCTACAGTGAAATCTCGATGGAGTGGGTCCCATCAGTTTGAACAGTTGTCTGGATCCATTTTGTGGATGGCACCAGAAGTCATCAGAATGCAAGATAAAAATCCATACAGCTTTCAGTCAGATGTATATGCATTTGGAATTGTTCTGTATGAATTGATGACTGGACAGTTACCTTATTCAAACATCAACAACAGGGACCAGATAATTTTTATGGTGGGACGAGGATACCTGTCTCCAGATCTCAGTAAGGTACGGAGTAACTGTCCAAAAGCCATGAAGAGATTAATGGCAGAGTGCCTCAAAAAGAAAAGAGATGAGAGACCACTCTTTCCCCAAATTCTCGCCTCTATTGAGCTGCTGGCCCGCTCATTGCCAAAAATTCACCGCAGTGCATCAGAACCCTCCTTGAATCGGGCTGGTTTCCAAACAGAGGATTTTAGTCTATATGCTTGTGCTTCTCCAAAAACACCCATCCAGGCAGGGGGATATGGTGCGTTTCCTGTCCACTGA',0.05)
# len(error)
```

**Creating the Mutated Genes**

```python
mutated_genes = pd.DataFrame(columns = ['GENE','ID_SAMPLE','MUTATED_SEQ'])
erroneous_seqs = pd.DataFrame(columns = ['GENE','ID_SAMPLE','ERRONEOUS_SEQ'])

# generating mutated gene sequences
def generate_mutated_seq(gene,samples_dataframe,actual_mutated_seq,preprocessed_data):
  print(len(samples_dataframe))
  print(len(actual_mutated_seq))
  print(gene)
  
  # getting unique samples to avoid generating duplicate sequences
  unique_samples=list(samples_dataframe[' ID_SAMPLE'].unique())
  # print(len(unique_samples))

  for sample in unique_samples:
    mutations_per_sample = samples_dataframe[samples_dataframe[' ID_SAMPLE'].isin([sample])]
    if len(mutations_per_sample) != 1:
      print('No of mutations for sample '+str(sample)+ ': '+str(len(mutations_per_sample)))     # samples having more than 1 mutation
      mutated_gene = gene_seq[gene]   # initializing the mutated_gene first with the original reference gene sequence
      mut_count = 1     # count to keep track of which mutation is being applied
      for index,record in mutations_per_sample.iterrows():
        if re.search('Substitution',record[' MUTATION_DESCRIPTION']):
          print('substitution mutation count '+str(mut_count))
          mutated_gene = get_mutated_gene_sequence_sub_multiple(record[' MUTATION_CDS'],mutated_gene)
          mut_count = mut_count + 1
      actual_mutated_seq = actual_mutated_seq.append({'GENE':gene,'ID_SAMPLE':sample,'MUTATED_SEQ':mutated_gene},ignore_index=True)
      # print(mutated_gene)
      preprocessed_data = preprocessed_data.append({'GENE':gene,'ID_SAMPLE':sample,'ERRONEOUS_SEQ':sim_error(mutated_gene,0.05)},ignore_index=True)
      mut_count = 0
    else:
      if re.search('Substitution',mutations_per_sample.iloc[0][' MUTATION_DESCRIPTION']):
        mutated_gene = get_mutated_gene_sequence_sub(gene,mutations_per_sample.iloc[0][' MUTATION_CDS'])
        # print(mutated_gene)
        actual_mutated_seq = actual_mutated_seq.append({'GENE':gene,'ID_SAMPLE':sample,'MUTATED_SEQ':mutated_gene},ignore_index=True)
        preprocessed_data = preprocessed_data.append({'GENE':gene,'ID_SAMPLE':sample,'ERRONEOUS_SEQ':sim_error(mutated_gene,0.05)},ignore_index=True)

  return actual_mutated_seq,preprocessed_data


mutated_genes,erroneous_seqs = generate_mutated_seq('BRAF',braf_mutated,mutated_genes,erroneous_seqs)
mutated_genes,erroneous_seqs = generate_mutated_seq('NRAS',nras_mutated,mutated_genes,erroneous_seqs)
mutated_genes,erroneous_seqs = generate_mutated_seq('MUC16',muc16_mutated,mutated_genes,erroneous_seqs)
mutated_genes,erroneous_seqs = generate_mutated_seq('LRP1B',lrp1b_mutated,mutated_genes,erroneous_seqs)
```
